## pi计算函数
### cal_pi_1 泰勒级数方法 输入n的大小
### cal_pi_2 蒙特卡洛方法 输入随机点的总量
### cal_pi_3 Chudnovsky方法 输入k的大小
### cal_pi_4 迭代算法 输入迭代次数

```python
import math
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cal_pi_1(n_max): #talyor
    pi=0
    for n in range(0, n_max+1):
        pi += (-1)**n/(2*n+1)
    print('the result of talyor is',pi*4)    
```

```python
def cal_pi_2(n_max): #monte carlo 蒙特卡洛
    n_in=0
    x_list = []
    y_list = []
    for i in range(n_max):
        x=random.random()
        y=random.random()
        x_list.append(x)
        y_list.append(y)
        if x**2+y**2 <= 1:
            n_in += 1
    
    plt.scatter(x_list, y_list,color = 'red', s = 1)
    x_c = np.linspace(0,1,100)
    y_c = np.sqrt(1 - x_c**2)
    plt.plot(x_c, y_c, color = 'blue', linewidth=1.0)
    plt.show()
    print('the result of monte carlo is',4*n_in/n_max)
```

```python
def cal_pi_3(k_max): #Chudnovsky
    down_elem = 0
    for k in range(k_max):
        a = math.factorial(6 * k) * (13591409 + 545140134 * k)
        b = math.factorial(3 * k) * (math.factorial(k) ** 3) * ((-640320) ** (3 * k))
        down_elem += a/b
    result = 426880 * math.sqrt(10005) / down_elem
    print('the result of Chudnovsky is',result)
```

```python
def cal_pi_4(times): #迭代算法
    a0 = 1
    b0 = 1/math.sqrt(2)
    t0 = 1/4
    p0 = 1
    for i in range(times):
        a = (a0 + b0) / 2
        b = math.sqrt(a0 * b0)
        t = t0 - p0 * ((a0 - a) ** 2)
        p = 2 * p0
        a0 = a
        b0 = b
        t0 = t
        p0 = p
    result = ((a + b) ** 2) / (4 * t)
    print('the result of iteration is',result)
```

```python
cal_pi_1(10**5)
cal_pi_2(3000)
cal_pi_3(100)
cal_pi_4(100)
```
