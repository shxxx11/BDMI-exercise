```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y_1 = np.sin(x)
y_2 = np.cos(x)
```

```python
plt.figure()
plt.subplot(211)
plt.plot(x, y_1, color = 'blue', linewidth=1.0)
plt.subplot(212)
plt.plot(x, y_2, color = 'blue', linewidth=1.0)
plt.show()
```
